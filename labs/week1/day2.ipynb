# Chat Compeletions API

```python
import requests

# Confirm if the Ollama service is active
requests.get("http://localhost:11434").content
```

```python
from openai import OpenAI

# Create client
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
```

```python
from IPython.display import Markdown, display

# Interact with the Ollama server via its OpenAI-compatible endpoint
response = ollama.chat.completions.create(
    model="gemma3:4b",
    messages=[
        {
            "role": "user",
            "content": "Tell me a fun fact about AI"
        }
    ]
)

# Format and display the response
display(Markdown(response.choices[0].message.content))
```

```python
# Check the json response
response.model_dump_json()
```
